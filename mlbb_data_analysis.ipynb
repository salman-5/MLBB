```python
import pyscipopt as scip

# Define the dimensions of the material to be cut
material_width = 100
material_height = 100

# Define the dimensions of each piece and the number of pieces of each size
pieces = [
    {'width': 20, 'height': 30, 'n_pieces': 5},
    {'width': 30, 'height': 40, 'n_pieces': 3},
    {'width': 40, 'height': 50, 'n_pieces': 2},
    {'width': 50, 'height': 60, 'n_pieces': 1}
]

# Create a SCIP solver instance
solver = scip.Solver()

# Disable output from the solver
solver.setIntParam('display/verblevel', 0)

# Create binary variables for each piece and orientation and for each position
# where a piece can be placed on the material
pieces_vars = {}
for i, piece in enumerate(pieces):
    for j in range(2):
        for x in range(material_width - piece['width' if j == 0 else 'height'] + 1):
            for y in range(material_height - piece['height' if j == 0 else 'width'] + 1):
                pieces_vars[(i, j, x, y)] = solver.addVar(name=f'piece_{i}_{j}_{x}_{y}', vtype='B')

# Define the objective function as the sum of the binary variables
solver.setObjective(solver.quicksum(pieces_vars.values()), sense='minimize')

# Define constraints to ensure that each piece is placed only once and that the placement is valid
for i, piece in enumerate(pieces):
    for j in range(2):
        for x in range(material_width - piece['width' if j == 0 else 'height'] + 1):
            for y in range(material_height - piece['height' if j == 0 else 'width'] + 1):
                solver.addCons(solver.quicksum(pieces_vars[(i, j, x2, y2)] for x2 in range(x, x + piece['width' if j == 0 else 'height']) for y2 in range(y, y + piece['height' if j == 0 else 'width'])) <= piece['n_pieces'])

# Define constraints to ensure that the material is not overlapped
for x in range(material_width):
    for y in range(material_height):
        solver.addCons(solver.quicksum(pieces_vars[(i, j, x2, y2)] for i in range(len(pieces)) for j in range(2) for x2 in range(x, x + pieces[i]['width' if j == 0 else 'height']) for y2 in range(y, y + pieces[i]['height' if j == 0 else 'width'])) <= 1)

# Solve the optimization problem
solver.optimize()

# Print the solution
for i, piece in enumerate(pieces):
    for j in range(2):
        for x in range(material_width):
            for y in range(material_height):
                if pieces_vars[(i, j, x, y)].getVal() == 1:
                    print(f"Piece {i}, orientation {j}, placed at ({x}, {y})")
```
